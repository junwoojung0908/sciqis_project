```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,t):
    return x * np.exp(-x*(1-t)) * np.sin(x+t) * (1-t)

def plot_function():
    x = np.linspace(0, 10, 100)
    tL = np.linspace(0, 0.4, 5)
    plt.figure(figsize=(10, 6))
    for t in tL:
        y = f(x, t)
        plt.plot(x, y, label=f't={t:.2f}')
    plt.title('Plot of f(x, t) for different values of t')
    plt.xlabel('x')
    plt.ylabel('f(x, t)')
    plt.legend()
    plt.grid()
    plt.show()

if __name__ == "__main__":
    plot_function()
```
